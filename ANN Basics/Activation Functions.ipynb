## **Activation Functions**
___

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step Function 
def step_function(x):
    return np.where(x >= 0, 1, 0)


x = np.linspace(-10, 10, 100)
y = step_function(x)

plt.plot(x, y)
plt.title("Step Function")
plt.ylim(-0.2, 1.2)
plt.grid(True)
plt.show()
```

```python
# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 100) # Generate 100 points between -10 and 10
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.show()

"""
max(0, x) vs maximum(0, x) with array input:
- max(0, x) is a built-in function that compares 0 with the one value of x and returns the maximum. It does not work element-wise on arrays.
- np.maximum(0, x) is a NumPy function that compares 0 with each element of the array x and returns an array of the maximum values. It works element-wise on arrays.
"""
```

```python
# Tanh Function
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.plot(x, y)
plt.title("Tanh Function")
plt.show()
```

```python
# Softmax Function
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # stability trick  for avoiding overflow
    return exp_x / np.sum(exp_x)

"""
[0.65900114 0.24243297 0.09856589]:
 means the first value (2.0) has the highest probability with approximately 65.9% so class 1 is the most likely class, 
 followed by 1.0 and then 0.1. The softmax function converts the input values into 
 probabilities that sum to 1, making it useful for multi-class classification problems.
"""
```

```python
x = np.array([2.0, 1.0, 0.1])
y = softmax(x)

labels = ["Class 1", "Class 2", "Class 3"]

plt.bar(labels, y)
plt.title("Softmax Output")
plt.ylim(0, 1)
plt.show()
"""
sofmax is not an activation function for hidden layers, 
it's typically used in the output layer of a neural network for multi-class classification problems.
It converts the raw output scores (logits) from the last layer into probabilities that sum to 1,
allowing us to interpret the output as class probabilities. For hidden layers,
we usually use activation functions like ReLU, sigmoid,
or tanh to introduce non-linearity and help the network learn complex patterns.
"""
```

```python
#Leaky ReLU Function
def leaky_relu(x):
    return np.where(x > 0, x, 0.1 * x)

x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

plt.plot(x, y)
plt.title("Leaky ReLU Function")
plt.grid(True)
plt.show()
```

| Function          | Shape                     | Range           | Main Use                    | Advantages                   | Disadvantages                                     |
| ----------------- | ------------------------- | --------------- | --------------------------- | ---------------------------- | ------------------------------------------------- |
| **Step Function** | Hard step                 | 0 or 1          | Perceptron (historical)     | Very simple                  | Not differentiable ❌, not usable in Deep Learning |
| **Sigmoid**       | S-shaped curve            | 0 → 1           | Binary classification       | Smooth, probabilistic output | Vanishing gradient problem ❌                      |
| **Tanh**          | S-shaped curve            | -1 → 1          | Hidden layers (sometimes)   | Zero-centered output         | Still suffers from vanishing gradient ❌           |
| **ReLU**          | Linear after 0            | 0 → ∞           | Hidden layers (most common) | Fast, simple, efficient 🔥   | Can cause “dead neurons” ❌                        |
| **Leaky ReLU**    | Small slope for negatives | (-∞ → ∞)        | Deep networks               | Solves dead neuron problem ✔ | Not always better than ReLU                       |
| **Softmax**       | Probability distribution  | 0 → 1 (sum = 1) | Output layer (multi-class)  | Produces probabilities       | Not used in hidden layers                         |
