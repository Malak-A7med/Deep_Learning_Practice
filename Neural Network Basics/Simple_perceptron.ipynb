## **Perceptron from Scratch**
___

```python
"""
In this notebook, we will implement a simple 
perceptron using Python and NumPy to understand 
how a basic neural unit works.
"""
```

```python
import numpy as np
```

```python
# Step function as the activation function
def step_function(x):
    return 1 if x >= 0 else 0
```

```python
class Perceptron:
    def __init__(self, input_size, lr=0.1):
        # Initialize weights and bias
        self.weights = np.zeros(input_size) # Initialize weights to zero >> [0, 0]
        self.bias = 0 # Initialize bias to zero
        self.learning_rate = lr 

    def predict(self, x):
        # Calculate the linear combination of inputs and weights
        linear_output = np.dot(x, self.weights) + self.bias

        # Apply the step function to get the predicted output
        predicted = step_function(linear_output)
        return predicted
    
    def train(self, x, y, epochs):
        for _ in range(epochs):
            for i in range(len(x)):
                # Make a prediction
                predicted = self.predict(x[i])

                # Update weights and bias based on the error
                error = y[i] - predicted
                self.weights += self.learning_rate * error * x[i]
                self.bias += self.learning_rate * error
```

```python
x = np.array([
    [0, 0], 
    [0, 1], 
    [1, 0], 
    [1, 1]]) # Input data for AND logic gate

y = np.array([0, 0, 0, 1]) # Output labels
```

```python
p = Perceptron(input_size=2)
p.train(x, y, epochs=10)
```

```python
# Test the perceptron
print(p.predict(np.array([0, 0]))) # Expected output: 0
print(p.predict(np.array([0, 1]))) # Expected output: 0
print(p.predict(np.array([1, 0]))) # Expected output: 0
print(p.predict(np.array([1, 1]))) # Expected output: 1
```

```python
for i in x:
    print(i, "->", p.predict(i))
```

```python
"""
- The perceptron learns a simple AND logic
- It adjusts weights based on errors
- It works only for linearly separable data
- it cannot learn XOR logic cuz it's not linearly separable
- For more complex patterns, we need multi-layer networks
"""
```

### **Try The Perceptron with XOR Gate**
___

```python
#try xor gate
x_xor = np.array([
    [0, 0], 
    [0, 1], 
    [1, 0], 
    [1, 1]]) # Input data for XOR logic gate

y_xor = np.array([0, 1, 1, 0]) # Output labels for XOR

p_xor = Perceptron(input_size=2)
p_xor.train(x_xor, y_xor, epochs=10)
for i in x_xor:
    print(i, "->", p_xor.predict(i))

# we see that the perceptron fails to learn XOR logic,
# cuz it's not linearly separable. This is a key limitation 
# of single-layer perceptrons, which is why we need 
# multi-layer networks to learn more complex patterns.
```
